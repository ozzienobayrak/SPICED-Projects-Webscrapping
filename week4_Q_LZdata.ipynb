```python
#WEB SCRAPPING
#Defining html
import requests
from bs4 import BeautifulSoup

url=requests.get('https://www.lyrics.com/artist/Led-Zeppelin/4739')
lyrics_html =url.text
lyrics_soup = BeautifulSoup(lyrics_html)
```

```python
lyrics_soup.head.title.text
```

```python
# #Find all links to songs of the artist on the webpage
lyrics_soup.find_all(class_="tal qx")
```

```python
lyrics_soup.find(class_="tal qx").a.text
```

```python
##Getting the title to lyrics
#title=lyrics_soup.body.find(class_="tal qx").a.text

##lyric-title

titles = []
all_titles = lyrics_soup.body.find_all(class_="tal qx")
for title in all_titles:
    titles.append(title.text)
titles
```

```python
#Getting the links as a list
base_url = 'https://www.lyrics.com/' #Define missing base of the url
#base_url+title
links = []
all_lyrics = lyrics_soup.find_all(class_="tal qx")
for lyric in all_lyrics:
    links.append(base_url + lyric.a['href'])
links
```

```python
#Getting the content of the links
import time
lyrics_contents = []
for link in links[:100]:
    #print(link)
    lyric_html = requests.get(link).text
    time.sleep(10)
    lyrics_LZ = BeautifulSoup(lyric_html)
    lyrics_contents.append(lyrics_LZ.find(class_="lyric-body").text.split('\n'))
    #print(lyrics_LZ.find(class_="lyric-body").text.split('\n'))
```

```python
lyrics_contents
```

```python
#Put them in a dataframe
import pandas as pd
df=pd.DataFrame({'title':titles[:100], 'links':links[:100], 'lyrics':lyrics_contents})
df['artist'] = 'Led Zeppelin'
```

```python
#QUEENS
url_q= requests.get('https://www.lyrics.com/artist/Queen/5205')
url_q.text[0:500]
```

```python
lyrics_q_html = url_q.text
lyrics_q_soup = BeautifulSoup(lyrics_q_html)
lyrics_q_soup.head.title.text
```

```python
#Getting the links as a list
links_q = []
all_lyrics_q = lyrics_q_soup.find_all(class_="tal qx")
for lyric in all_lyrics_q:
    links_q.append(base_url + lyric.a['href'])
links_q
```

```python
##Getting the title to lyrics
#title=lyrics_soup.body.find(class_="tal qx").a.text

##lyric-title

titles_q = []
all_titles_q = lyrics_q_soup.body.find_all(class_="tal qx")
for title in all_titles_q:
    titles_q.append(title.text)
titles_q
```

```python
#To extract the lyrics in the links:
import time
lyricsQ_contents = []
for link_q in links_q[:100]:
    lyric_q_html = requests.get(link_q).text
    time.sleep(10)
    lyrics_Q = BeautifulSoup(lyric_q_html)
    lyricsQ_contents.append(lyrics_Q.find(class_="lyric-body").text.split('\n'))
    #print(lyrics_LZ.find(class_="lyric-body").text.split('\n'))
```

```python
#Put them in a dataframe together
df_Q=pd.DataFrame({'title': titles_q[:100], 'links':links_q[:100], 'lyrics':lyricsQ_contents})
df_Q['artist'] = 'Queen'
df_all = df.append(df_Q)
```
